```python
import cv2
import numpy as np
import mediapipe as mp

mp_hands = mp.solutions.hands
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

xi = np.random.randint(200, 1000)
yi = 0

if xi > 600:
    x_fin = np.random.randint(20, 200)
else:
    x_fin = np.random.randint(1000, 1260)

slope = float(360 / (x_fin - xi))
vert = 'down'
skips = 10
x_cord = xi

# 벽돌 정보 생성
brick_rows = 8  # 변경: 벽돌 행 수 증가
brick_cols = 10
brick_width = 80
brick_height = 20
brick_gap = 10  # 변경: 벽돌 간격 설정
bricks = []
for i in range(brick_rows):
    for j in range(brick_cols):
        brick_x = j * (brick_width + brick_gap)
        brick_y = i * (brick_height + brick_gap)
        bricks.append([brick_x, brick_y, brick_width, brick_height])

with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=1,
    min_detection_confidence=0.5) as hands:

    while cap.isOpened():
        _, frame = cap.read()
        frame = cv2.flip(frame, 1)
        results = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

        # 벽돌 그리기
        for brick in bricks:
            cv2.rectangle(frame, (brick[0], brick[1]), (brick[0] + brick[2], brick[1] + brick[3]), (0, 255, 0), -1)

        if vert == 'down':
            yi += skips
        else:
            yi -= skips
        xi += skips / slope

        if yi < 20:
            vert = 'down'
        if yi > 360:
            vert = 'up'
        if xi < 15 or xi > 1260:
            slope *= -1

        # 공 그리기
        frame = cv2.circle(frame, (int(xi), yi), 15, (0, 255, 0), -1)

        if results.multi_hand_landmarks:
            hn = results.multi_hand_landmarks[0]
            x_cord = int(hn.landmark[8].x * frame.shape[1])
            x_cord = min(1279 - 70, x_cord)
            x_cord = max(70, x_cord)

        # 패들 그리기
        frame = cv2.rectangle(frame, (x_cord - 70, 375), (x_cord + 70, 425), (0, 0, 255), -1)
        if (yi>=360) and (not(xi>x_cord-120) and (xi <x_cord+120)):
            print('Lost')
            break
        # 게임 로직: 공과 벽돌의 충돌 체크
        for brick in bricks:
            if yi <= brick[1] + brick[3] and xi >= brick[0] and xi <= brick[0] + brick[2]:
                bricks.remove(brick)
                slope *= -1

        cv2.imshow('window', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
