#### CHATGPT를 활용한 pygame으로 벽돌깨기 게임 만들어보기
- 기능을 추가요청을 반복후 달린 주석을 바탕으로 수정

```python
# !pip install pygame
!pip show pygame
```

```python
import pygame
import sys
import random

# Pygame 초기화
pygame.init()

# 화면 설정
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("공 맞추기")

# 색상
color = {
    'white': (255, 255, 255),
    'black': (0, 0, 0),
    'red': (255, 0, 0)
}

# 공 속성
ball_radius = [5, 20, 200, 1, 100, 5, 20]
ball_radius = random.choice(ball_radius)
pygame.display.set_caption('사이즈 %d: 공 맞추기 게임' % ball_radius)

ball_x = random.randint(ball_radius, width - ball_radius)
ball_y = random.randint(ball_radius, height )
ball_vel_x = random.choice([-5, 5])
ball_vel_y = random.choice([-5, 5])

# 바닥 막대 속성
paddle_width, paddle_height = 150, 20
paddle_x = width // 2 - paddle_width // 2
paddle_y = height - paddle_height - 10
paddle_vel = 7

# 벽돌 속성 - 다층으로 벽돌 생성
brick_width, brick_height = 70, 50
brick_cols = 10
brick_rows = 3  # 초기 벽돌 층 수 변경
bricks = []
for row in range(brick_rows):
    for i in range(brick_cols):
        brick_x = i * (brick_width + 5)
        brick_y = row * (brick_height + 5) + 50  # 여러 층을 위한 y 위치 설정
        bricks.append(pygame.Rect(brick_x, brick_y, brick_width, brick_height))

clock = pygame.time.Clock()

# 점수
score = 0
font = pygame.font.Font(None, 36)

# 현재 스테이지
current_stage = 1

# 게임 루프
running = True
while running:
    screen.fill(color['white'])

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_z:
                ball_radius = random.randint(10, 30)

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and paddle_x > 0:
        paddle_x -= paddle_vel
    if keys[pygame.K_RIGHT] and paddle_x < width - paddle_width:
        paddle_x += paddle_vel

    ball_x += ball_vel_x
    ball_y += ball_vel_y

    if ball_x + ball_radius >= width or ball_x - ball_radius <= 0:
        ball_vel_x *= -1
    if ball_y - ball_radius <= 0:
        ball_vel_y *= -1
    elif ball_y + ball_radius >= height:
        ball_vel_y *= -1
        if not (paddle_x <= ball_x <= paddle_x + paddle_width):
            running = False

    ball_rect = pygame.Rect(ball_x - ball_radius, ball_y - ball_radius, ball_radius * 2, ball_radius * 2)

    for brick in bricks:
        if brick.colliderect(ball_rect):
            ball_vel_y *= -1
            bricks.remove(brick)
            score += 1

    for brick in bricks:
        pygame.draw.rect(screen, color['black'], brick)

    pygame.draw.circle(screen, color['red'], (ball_x, ball_y), ball_radius)
    pygame.draw.rect(screen, color['black'], (paddle_x, paddle_y, paddle_width, paddle_height))

    score_text = font.render("Score: " + str(score), True, color['black'])
    screen.blit(score_text, (10, 10))

    # 모든 블록이 깨졌을 때 다음 스테이지로 넘어가기
    if len(bricks) == 0:
        current_stage += 1
        bricks = []  # 기존의 벽돌을 모두 제거
        for row in range(brick_rows + current_stage):  # 다음 스테이지 벽돌 생성
            for i in range(brick_cols):
                brick_x = i * (brick_width + 5)
                brick_y = row * (brick_height + 5) + 50
                bricks.append(pygame.Rect(brick_x, brick_y, brick_width, brick_height))

    pygame.display.update()
    clock.tick(60)

pygame.quit()
sys.exit()
```
