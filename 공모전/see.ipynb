```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# 파일 경로
file_path = '장애인 취업정보 자치구 뽑아낸거.csv'
geojson_path = 'seoul_municipalities_geo_simple.json'

# 데이터 로드
data = pd.read_csv(file_path)

# 자치구별 취업자 수 집계
borough_counts = data['자치구'].value_counts()
borough_counts_df = borough_counts.reset_index()
borough_counts_df.columns = ['자치구', '취업자 수']

# 서울시 자치구 지리적 경계 데이터 로드
seoul_map = gpd.read_file(geojson_path)

# 데이터 병합
merged_data = seoul_map.set_index('name').join(borough_counts_df.set_index('자치구'))
merged_data['취업자 수'] = merged_data['취업자 수'].fillna(0)

# 시각화
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged_data.plot(column='취업자 수', ax=ax, legend=True, cmap='OrRd', edgecolor='black')
ax.set_title('서울시 자치구별 장애인 취업자 수', fontsize=15)

# 자치구 이름 추가
for idx, row in merged_data.iterrows():
    plt.annotate(text=idx, xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', fontsize=8)

ax.axis('off')
plt.savefig('시각화/자치구별 장애인 취업자 수.png', format='png',bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 차량 빈도 데이터 로드
vehicle_frequency_df = pd.read_csv('서울시 21년도 차량빈도 구,평균.csv', encoding='CP949')

# 사회복지시설 데이터 로드
social_welfare_facilities_df = pd.read_csv('서울시 사회복지시설 목록.csv', encoding='CP949')

# 사회복지시설 수 집계
social_welfare_facilities_df['시군구명'] = social_welfare_facilities_df['시군구명'].str.strip()
facilities_count = social_welfare_facilities_df['시군구명'].value_counts().reset_index()
facilities_count.columns = ['구', '사회복지시설 수']

# 차량 빈도 데이터에 대한 구별 코드를 구 이름으로 매핑
vehicle_frequency_df['구'] = vehicle_frequency_df['구'].map({
    11010: '종로구', 11020: '중구', 11030: '용산구', 11040: '성동구', 11050: '광진구',
    11060: '동대문구', 11070: '중랑구', 11080: '성북구', 11090: '강북구', 11100: '도봉구',
    11110: '노원구', 11120: '은평구', 11130: '서대문구', 11140: '마포구', 11150: '양천구',
    11160: '강서구', 11170: '구로구', 11180: '금천구', 11190: '영등포구', 11200: '동작구',
    11210: '관악구', 11220: '서초구', 11230: '강남구', 11240: '송파구', 11250: '강동구'
})


# 데이터 병합
merged_data = pd.merge(vehicle_frequency_df, facilities_count, on='구', how='left')
merged_data['사회복지시설 수'] = merged_data['사회복지시설 수'].fillna(0)

# 데이터 정규화
merged_data['평균 차량 빈도 정규화'] = (merged_data['평균'] - merged_data['평균'].min()) / (merged_data['평균'].max() - merged_data['평균'].min())
merged_data['사회복지시설 수 정규화'] = (merged_data['사회복지시설 수'] - merged_data['사회복지시설 수'].min()) / (merged_data['사회복지시설 수'].max() - merged_data['사회복지시설 수'].min())

# 센서 설치 우선 순위 점수 계산
merged_data['센서 설치 우선 순위 점수'] = 0.5 * merged_data['평균 차량 빈도 정규화'] + 0.5 * merged_data['사회복지시설 수 정규화']

# 데이터 정렬
merged_data_sorted = merged_data.sort_values(by='센서 설치 우선 순위 점수', ascending=False)

# 시각화
plt.figure(figsize=(14, 8))
sns.barplot(x='구', y='센서 설치 우선 순위 점수', data=merged_data_sorted, palette='coolwarm')
plt.title('서울시 구별 장애인 불법 주차 감지 센서 설치 우선 순위', fontsize=15)
plt.xlabel('구', fontsize=12)
plt.ylabel('센서 설치 우선 순위 점수', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

# 서울시 자치구별 지리적 경계 데이터 로드
geo_data_path = 'seoul_municipalities_geo_simple.json'
seoul_geo_df = gpd.read_file(geo_data_path)

# 데이터 준비 및 정규화
merged_data['평균 차량 빈도 정규화'] = (merged_data['평균'] - merged_data['평균'].min()) / (merged_data['평균'].max() - merged_data['평균'].min())
merged_data['사회복지시설 수 정규화'] = (merged_data['사회복지시설 수'] - merged_data['사회복지시설 수'].min()) / (merged_data['사회복지시설 수'].max() - merged_data['사회복지시설 수'].min())
X = merged_data[['평균 차량 빈도 정규화', '사회복지시설 수 정규화']].values

# K-평균 군집화 수행 (클러스터 수 = 3으로 설정)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(X)

# 클러스터 레이블을 데이터프레임에 추가
merged_data['Cluster'] = cluster_labels

# 자치구 중심 좌표 추출
seoul_geo_df['center'] = seoul_geo_df['geometry'].centroid
centers = seoul_geo_df.set_index('name')['center']

# 군집화 결과와 지리적 데이터 병합
merged_geo_data = seoul_geo_df.set_index('name').join(merged_data.set_index('구'))

# 지도에 군집화 결과 시각화
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
seoul_geo_df.plot(ax=ax, color='white', edgecolor='black')
scatter = ax.scatter(centers.x, centers.y, c=merged_geo_data['Cluster'], cmap='viridis', s=100)

# 각 자치구 이름 표시
for idx, point in centers.items():
    ax.text(point.x, point.y, idx, fontsize=12, ha='center')

# 범례 추가
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)

ax.set_title('서울시 자치구별 장애인 불법 주차 감지 센서 설치 군집화 결과', fontsize=15)
ax.axis('off')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import json

# CSV 파일 로드 및 데이터 처리
file_path = '주차장(동별)(2016년+이후)_20240118202906.csv'
parking_data = pd.read_csv(file_path)

# 주차장 수를 나타내는 열 식별 및 집계
count_columns = [col for col in parking_data.columns if "개소 (개소)" in str(parking_data[col].iloc[2])]
aggregated_counts = parking_data.iloc[4:][['동별(2)'] + count_columns]
aggregated_counts[count_columns] = aggregated_counts[count_columns].apply(pd.to_numeric, errors='coerce')
aggregated_counts['Total_Parking_Lots'] = aggregated_counts[count_columns].sum(axis=1)

# GeoJSON 파일 로드
geojson_path = 'seoul_municipalities_geo_simple.json'
with open(geojson_path, 'r', encoding='utf-8') as file:
    geo_data = json.load(file)

# GeoJSON 데이터를 GeoDataFrame으로 변환
gdf = gpd.GeoDataFrame.from_features(geo_data['features'])

# 주차장 데이터와 지리 데이터 결합
merged_data = gdf.merge(aggregated_counts.rename(columns={'동별(2)': 'name'}), left_on='name', right_on='name')

# 데이터 시각화
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged_data.plot(column='Total_Parking_Lots', ax=ax, legend=True, cmap='OrRd')

# 자치구 이름 표시
for idx, row in merged_data.iterrows():
    plt.annotate(text=row['name'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', verticalalignment='center', fontsize=8,color='black')

plt.title('서울시 주차장수 분포')
plt.savefig('시각화/서울시 주차장수 분포.png', format='png',bbox_inches='tight')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# 서울시 자치구 경계 데이터 파일 경로
seoul_geojson_path = 'seoul_municipalities_geo_simple.json'

# GeoJSON 데이터 불러오기
seoul_geo_df = gpd.read_file(seoul_geojson_path)

# 차량 빈도 데이터 파일 경로
traffic_freq_path = "서울시 21년 차량빈도_서울만.csv"

# 차량 빈도 데이터 불러오기
traffic_freq_df = pd.read_csv(traffic_freq_path, encoding='utf-8-sig')

# 자치구별로 평균 차량 빈도를 계산
traffic_freq_avg = traffic_freq_df.groupby('내용')['FRIN_CG'].mean().reset_index()

# 자치구 이름을 GeoJSON 데이터의 'name' 컬럼과 일치시키기 위해 컬럼명 변경
traffic_freq_avg.rename(columns={'내용': 'name', 'FRIN_CG': 'avg_traffic'}, inplace=True)

# GeoJSON 데이터와 차량 빈도 데이터 결합
seoul_traffic_geo_df = seoul_geo_df.merge(traffic_freq_avg, on='name', how='left')



# 차량 빈도에 따른 시각화 (한글 폰트 적용)
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
seoul_traffic_geo_df.plot(column='avg_traffic', ax=ax, legend=True,
                          legend_kwds={'label': "자치구별 평균 차량 빈도"},
                          cmap='OrRd', edgecolor='black')

# 자치구 이름 표시 (한글 폰트 적용)
for idx, row in seoul_traffic_geo_df.iterrows():
    plt.annotate(text=row['name'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', fontsize=9)

# 지도에 제목 추가
plt.title('서울시 자치구별 평균 차량 빈도 시각화', fontsize=15)

# 축 레이블 제거
ax.set_axis_off()
plt.savefig('시각화/자치구별 평균 차량 빈도.png', format='png',bbox_inches='tight')
# 시각화 결과 출력
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

def forward_selection(data, response):
    remaining = set(data.columns)
    remaining.remove(response)
    selected = []
    current_score, best_new_score = float('inf'), float('inf')
    
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = "{} ~ {}".format(response, ' + '.join(selected + [candidate]))
            score = sm.OLS.from_formula(formula, data).fit().aic
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop(0)
        if current_score > best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
    formula = "{} ~ {}".format(response, ' + '.join(selected))
    model = sm.OLS.from_formula(formula, data).fit()
    return model

# Load the data
data = pd.read_csv('최종.csv')  # Replace 'file_path' with your data file path

# Apply forward selection
model = forward_selection(data, '과태료건수')

# Print the model summary
print(model.summary())
```

```python
# 불용어 처리
stop_words = ['장애인' , '장애인들'  , '그동안' , '편의 시설' , '편의' , '시설' , '관계자' , '사업비' , '만큼' , '위원회' ,'비장애인' ,'무장애','a씨','b씨']

key_word = [word for word in key_word if not word in stop_words]
print(len(key_word))
```

```python
# 가장 많이 나온 키워드 (상위 60개)
counts = Counter(key_word)
tags = counts.most_common(60)
print(tags)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from itertools import combinations
from collections import Counter
import pandas as pd

# 데이터 파일 로드
news_data = pd.read_excel('뉴스.xlsx')

# '키워드' 컬럼에서 각 뉴스 아이템의 키워드를 분리
keyword_lists = news_data['키워드'].dropna().str.split(',')

# 각 뉴스 아이템의 키워드 쌍 생성
keyword_pairs = [list(combinations(sublist, 2)) for sublist in keyword_lists if len(sublist) > 1]

# 키워드 쌍 리스트 평탄화
flat_pairs = [pair for sublist in keyword_pairs for pair in sublist]

# "장애인"과 "주차"와 관련된 키워드에 대한 필터링
related_keywords = ['장애인', '주차']
filtered_keyword_pairs = [pair for pair in flat_pairs if any(keyword in pair for keyword in related_keywords)]

# 각 키워드 쌍의 발생 횟수 카운팅
filtered_pair_counts = Counter(filtered_keyword_pairs)

# 상위 10개 키워드 추출
top_10_keywords = Counter([word for pair in filtered_pair_counts.keys() for word in pair]).most_common(10)
top_10_keywords = [keyword for keyword, _ in top_10_keywords]

# 상위 10개 키워드에 해당하는 쌍만 필터링
top_10_pairs = [pair for pair in filtered_keyword_pairs if pair[0] in top_10_keywords and pair[1] in top_10_keywords]

# 각 쌍의 발생 횟수 카운팅
top_10_pair_counts = Counter(top_10_pairs)

# 상위 10개 키워드만 포함하는 그래프 생성
G_top_10 = nx.Graph()
for pair, count in top_10_pair_counts.items():
    G_top_10.add_edge(pair[0], pair[1], weight=count)

# 그래프 그리기
plt.figure(figsize=(8, 8))
pos_top_10 = nx.spring_layout(G_top_10, k=0.5)
nx.draw_networkx_nodes(G_top_10, pos_top_10, node_size=100)
nx.draw_networkx_edges(G_top_10, pos_top_10, alpha=0.7, width=[G_top_10[u][v]['weight'] / 2 for u, v in G_top_10.edges()])
nx.draw_networkx_labels(G_top_10, pos_top_10, font_size=12)
plt.title("Top 10 Keyword Association Graph")
plt.axis('off')
plt.show()
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# 파일 경로
file_path = '뉴스.xlsx'

# 파일 읽기
news_data = pd.read_excel(file_path)

# '특성추출(가중치순 상위 50개)' 컬럼에서 모든 키워드 추출
all_keywords = []
for keywords in news_data['특성추출(가중치순 상위 50개)']:
    all_keywords.extend(keywords.split(','))

# 키워드 빈도수 계산
keyword_counts = Counter(all_keywords)

# 가장 많이 등장하는 상위 50개 키워드 추출
top_50_keywords = keyword_counts.most_common(50)

# 결과를 DataFrame으로 변환
top_50_df = pd.DataFrame(top_50_keywords, columns=['Keyword', 'Frequency'])

# 그래프 그리기
plt.figure(figsize=(10, 8))
plt.barh(y=[keyword for keyword, _ in top_50_keywords], 
         width=[count for _, count in top_50_keywords])
plt.gca().invert_yaxis()  # 높은 값이 상단에 오도록 y축 반전
plt.xlabel('Frequency')
plt.ylabel('Keyword')
plt.title('Top 50 Keywords in News Data')
plt.tight_layout()

# 그래프 저장
graph_path = 'graph.png'
plt.savefig(graph_path)

# 결과 확인
print(top_50_df.head())  # 상위 5개만 표시
```

```python
import seaborn as sns
top_50_df = pd.DataFrame(top_50_keywords, columns=['Keyword', 'Frequency'])

# 그래프 스타일 설정
sns.set_style("whitegrid")

# 그래프 그리기
plt.figure(figsize=(12, 10))
barplot = sns.barplot(x='Frequency', y='Keyword', data=top_50_df, palette='viridis')
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Keyword', fontsize=14)
plt.title('뉴스 키워드', fontsize=16)

# 각 바에 빈도수 텍스트 추가
for p in barplot.patches:
    width = p.get_width()
    plt.text(width + 1, p.get_y() + p.get_height()/2. + 0.2, '{:1.0f}'.format(width), ha="left")

# 그래프 저장
graph_path = 'improved_graph.png'
plt.savefig(graph_path)

# 결과 확인
print(top_50_df.head())  # 상위 5개만 표시
```
