```python
import cv2
import mediapipe as mp

cap = cv2.VideoCapture(0)
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_draw = mp.solutions.drawing_utils

ball_x, ball_y = 300, 300
ball_radius = 20
ball_speed_x, ball_speed_y = 0, 0  # 공의 초기 속도는 0으로 설정

prev_thumb_x, prev_thumb_y = 0, 0
prev_pinky_x, prev_pinky_y = 0, 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    img_RGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(img_RGB)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            thumb = hand_landmarks.landmark[8]  # Thumb landmark
            pinky = hand_landmarks.landmark[12]  # Pinky landmark
            thumb_x, thumb_y = int(thumb.x * frame.shape[1]), int(thumb.y * frame.shape[0])
            pinky_x, pinky_y = int(pinky.x * frame.shape[1]), int(pinky.y * frame.shape[0])

            cv2.circle(frame, (thumb_x, thumb_y), 10, (255, 0, 0), -1)
            cv2.circle(frame, (pinky_x, pinky_y), 10, (0, 255, 0), -1)

            if prev_thumb_x != 0 and prev_thumb_y != 0 and prev_pinky_x != 0 and prev_pinky_y != 0:
                cv2.line(frame, (prev_thumb_x, prev_thumb_y), (prev_pinky_x, prev_pinky_y), (0, 0, 255), 3)

                # Calculate the direction of the line (thumb to pinky)
                direction_x = thumb_x - pinky_x
                direction_y = thumb_y - pinky_y

                # Update the ball's speed according to the direction of the line
                ball_speed_x = direction_x // 20
                ball_speed_y = direction_y // 20

            # Apply the force to the ball
            ball_x += ball_speed_x
            ball_y += ball_speed_y

            # Boundaries for the ball to stay within the screen
            if ball_x - ball_radius <= 0 or ball_x + ball_radius >= frame.shape[1]:
                ball_speed_x *= -1
            if ball_y - ball_radius <= 0 or ball_y + ball_radius >= frame.shape[0]:
                ball_speed_y *= -1

            # Draw the ball
            cv2.circle(frame, (ball_x, ball_y), ball_radius, (255, 255, 0), -1)

            prev_thumb_x, prev_thumb_y = thumb_x, thumb_y
            prev_pinky_x, prev_pinky_y = pinky_x, pinky_y

    cv2.imshow('Hand Tracking', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
