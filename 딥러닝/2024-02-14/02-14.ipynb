#### 구글 검색 - 활성화함수


Sigmoid 
- 결과가 0에서 1 사이만 결과가 나오게끔 한다
- (x 부분이 wx+b임)
- 레이어를 거치면 거칠수록 값이 현저하게 작아지게 되어 기울기 소실 현상이 발생




ReLU(Rectified Linear Unit)(정류한 선형 유닛)
- 이 활성화 함수가 시작하면서 암흑기가 풀리게 되었다.
- +/-가 반복되는 신호에서 -흐름을 차단한다는 의미
- 은닉층에서 굉장히 많이 사용
- 양수면 자기자신 반환, 음수면 0을 반환
- 단순한 구조로 경사 
- 단점
  - 음수 값이 모두 0으로 반환되므로 기울기도 모조리 0이됨(입력값이 음수이면)
  - 가중치 업데이트 안되는 현상 발생 가능

### y값 분류

<span style='color: #ffd33d'>다항(출력값이 3개이상)은 무조건 원핫 인코딩 softmax</span>



##### y표정 분류
- 슬픔 [0,0,1]  <--- [0.3,0.2,0.5] 시그모이드 한값을 합해서 나누기 (합이 1) <--- [0.3,0.5,0.8] 시그모이드 (wx+b)
- 기쁨 [1,0,0]  <--- [0.7,0.1,0.2]  
- 분노 [0,1,0]
- 슬픔 [0,0,1]

softmax함수는 wx+b 한값을 sigmoid 해서 0~1 값으로 변경한 여러개의 값을 더해서 각 값을 나눈 확률값




이항분류는 사용자의 선택에 따라 시그모이드 또는 softmax

- y값이 출력갯수가 1개이면 sigmoid 
- y값의 출력갯수가 2개이면 softmax


||sigmoid|softmax|
|------|---|---|
|웃음|0|[1,0]|
|울음|1|[0,1]|
