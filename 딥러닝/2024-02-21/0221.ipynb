lr 이 커지면 w,b 갱신 폭이 커진다.
- 즉 이전의 w값이 3이였다면 -> 
  - lr 값이 0.1이면 w 값이 이후에 0.7이라면
  - lr 값이 100이면 w값이 이후에 700일 수 있다(단순가정)
- 즉 wx+b 함으로 이전데이터와 값차이가 lr값이 0.1보다 100이 훨씬 더 많이 생긴다
