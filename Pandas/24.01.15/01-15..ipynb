#### 스피어만 상관계수 (corr)로 부르는 거랑 날코딩

```python
import numpy as np

# 두 변수의 데이터 생성
x = np.array([1, 2, 3, 4, 5])
y = np.array([-1, -2,- 3, -4, -5])

# Pearson 상관계수 계산
correlation_coefficient = np.corrcoef(x, y)[0, 1]

print("Pearson 상관계수:", correlation_coefficient)
```

```python
import numpy as np

# 두 변수의 데이터 생성
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])

# Pearson 상관계수 계산 함수
def calculate_pearson_correlation(x, y):
    n = len(x)
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator_x = np.sum((x - mean_x)**2)
    denominator_y = np.sum((y - mean_y)**2)
    correlation_coefficient = numerator / np.sqrt(denominator_x * denominator_y)
    return correlation_coefficient

# Pearson 상관계수 계산
result = calculate_pearson_correlation(x, y)

# 결과 출력
print("Pearson 상관계수:", result)
```

# 자료 세분화하는 방법

자료 세분화는 큰 데이터셋을 더 작고 관리하기 쉬운 부분으로 나누는 프로세스를 의미합니다. 이는 데이터를 이해하고 분석하는 데 도움이 되며, 특정 패턴이나 관계를 찾거나 머신러닝 모델에 적용하기 전에 데이터를 사전 처리하는 데 사용됩니다. 다양한 방법으로 자료를 세분화할 수 있습니다. 몇 가지 주요한 방법을 소개하겠습니다.

## 1. 범주화 (Binning 또는 Bucketing)
- 연속형 데이터를 구간으로 나누어 범주화합니다.
- 나이를 10대, 20대, 30대 등으로 그룹화하거나 수치형 데이터를 특정 구간으로 나눌 수 있습니다.
- 히스토그램 생성이나 머신러닝 모델에 적용하기 전에 데이터를 단순화하는 데 유용합니다.

## 2. 원핫 인코딩 (One-Hot Encoding)
- 범주형 변수를 이진 형태로 변환합니다.
- "색상" 변수가 있을 때, 빨강, 파랑, 녹색 등의 값은 0 또는 1로 표현됩니다.

## 3. 시계열 분해 (Time Series Decomposition)
- 시계열 데이터를 추세, 계절성 및 불규칙한 구성 요소로 나눕니다.
- 추세 분석과 계절성 요소를 이해하거나 예측 모델에 적용하기 위해 사용됩니다.

## 4. 차원 축소 (Dimensionality Reduction)
- 데이터의 차원을 축소하여 중요한 특징만을 유지합니다.
- PCA, t-SNE 등의 차원 축소 기법을 사용할 수 있습니다.

## 5. 군집화 (Clustering)
- 유사한 패턴이나 특성을 갖는 데이터를 그룹화합니다.
- K-Means, 계층적 군집화 등의 알고리즘을 사용할 수 있습니다.

## 6. 샘플링 (Sampling)
- 대규모 데이터셋에서 일부를 추출하여 작은 샘플로 만듭니다.
- 모델 훈련 및 테스트를 위해 데이터를 준비할 때 유용합니다.

## 7. 특성 선택 (Feature Selection)
- 중요한 특성만 선택하고 나머지는 제거합니다.
- 머신러닝 모델의 성능을 향상시키거나 계산 비용을 줄이기 위해 사용됩니다.

자료 세분화는 데이터의 특성과 목적에 따라 선택되어야 하며, 주어진 문제나 분석 목표에 따라 다양한 방법을 조합하여 사용할 수 있습니다.

## K-means 아래로 갈수록 날코딩

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 랜덤한 데이터 생성
np.random.seed(42)
X = np.concatenate([np.random.normal(0, 1, (100, 2)),
                    np.random.normal(4, 1, (100, 2)),
                    np.random.normal(8, 1, (100, 2))])

# K-Means 모델 생성 및 학습
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# 클러스터 중심과 클러스터 할당 결과 가져오기
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def kmeans(X, n_clusters=3, max_iters=100):
    # 초기 클러스터 중심 무작위 선택
    centers = X[np.random.choice(len(X), n_clusters, replace=False)]

    for _ in range(max_iters):
        # 각 데이터 포인트를 가장 가까운 클러스터에 할당
        labels = np.argmin(np.linalg.norm(X - centers[:, np.newaxis], axis=2), axis=0)

        # 새로운 클러스터 중심 계산
        new_centers = np.array([X[labels == k].mean(axis=0) for k in range(n_clusters)])

        # 중심 업데이트
        if np.all(centers == new_centers):
            break
        centers = new_centers

    return labels, centers

# 랜덤한 데이터 생성
np.random.seed(42)
X = np.concatenate([np.random.normal(0, 1, (100, 2)),
                    np.random.normal(4, 1, (100, 2)),
                    np.random.normal(8, 1, (100, 2))])

# K-Means 클러스터링 수행
labels, centers = kmeans(X, n_clusters=3)

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)
plt.title('K-Means Clustering (Custom Implementation)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def kmeans_plusplus(X, n_clusters=3, max_iters=100):
    # 첫 번째 중심점을 무작위로 선택
    centers = [X[np.random.choice(len(X))]]

    for _ in range(1, n_clusters):
        # 각 데이터 포인트와 현재 중심점들과의 최소 거리 제곱 계산
        distances = np.array([min(np.linalg.norm(center - x) ** 2 for center in centers) for x in X])
        
        # 확률 분포로부터 다음 중심점 선택
        prob_dist = distances / np.sum(distances)
        new_center_index = np.random.choice(len(X), p=prob_dist)
        centers.append(X[new_center_index])

    for _ in range(max_iters):
        # 각 데이터 포인트를 가장 가까운 클러스터에 할당
        labels = np.argmin(np.linalg.norm(X - np.array(centers)[:, np.newaxis], axis=2), axis=0)

        # 새로운 클러스터 중심 계산
        new_centers = np.array([X[labels == k].mean(axis=0) for k in range(n_clusters)])

        # 중심 업데이트
        if np.all(centers == new_centers):
            break
        centers = new_centers

    return labels, centers

# 랜덤한 데이터 생성
np.random.seed(42)
X = np.concatenate([np.random.normal(0, 1, (100, 2)),
                    np.random.normal(4, 1, (100, 2)),
                    np.random.normal(8, 1, (100, 2))])

# K-Means 클러스터링 (K-Means++ 적용)
labels, centers = kmeans_plusplus(X, n_clusters=3)

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(np.array(centers)[:, 0], np.array(centers)[:, 1], c='red', marker='X', s=200)
plt.title('K-Means Clustering with K-Means++ Initialization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
