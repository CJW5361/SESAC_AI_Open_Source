```python
import cv2
import numpy as np

# 흑백 공간 생성
height, width = 400, 500
img = np.zeros((height, width, 3), dtype=np.uint8)

def draw_shape(event, x, y, flags, param):
    if event == cv2.EVENT_RBUTTONDOWN:  # 마우스 오른쪽 버튼 클릭 감지
        cv2.rectangle(img, (x - 20, y - 20), (x + 20, y + 20), (255, 0, 0), 2)  # 파란색 사각형 그리기
    elif event == cv2.EVENT_LBUTTONDOWN:  # 마우스 왼쪽 버튼 클릭 감지
        cv2.circle(img, (x, y), 20, (0, 255, 255), 2)  # 금색(노랑+빨강) 원 테두리 그리기


cv2.namedWindow('Draw Circle')  # 창 생성
cv2.setMouseCallback('Draw Circle', draw_circle)  # 마우스 이벤트 콜백 설정

while True:
    cv2.imshow('Draw Circle', img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):  
        break

cv2.destroyAllWindows()
```

```python
import cv2

def apply_mosaic(image_path, x, y, width, height):
    # 이미지 읽기
    image = cv2.imread(image_path)

    # 모자이크 처리할 영역 선택
    selected_area = image[y:y+height, x:x+width]

    # 이미지 축소
    small_area = cv2.resize(selected_area, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_LINEAR)
    # 이미지 확대
    blurred_area = cv2.resize(small_area, (width, height), interpolation=cv2.INTER_NEAREST)

    # 원본 이미지에 모자이크 적용
    image[y:y+height, x:x+width] = blurred_area

    # 결과 이미지 출력
    cv2.imshow('Mosaic Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 함수 호출하여 모자이크 적용
apply_mosaic('dog.jpg', 400, 450, 500, 200)
```

```python
# 얼굴 정면 아니면 음성으로 '얼굴을 정면으로 해주세요' 음성발생
import cv2
import pyttsx3

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)

cap = cv2.VideoCapture(0)

# 음성 출력을 위한 초기화
engine = pyttsx3.init()

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # 얼굴 감지 여부 확인
    if len(faces) == 0:
        # 얼굴이 감지되지 않았을 때 음성 안내
        engine.say("얼굴이 감지되지 않았습니다. 정면을 향해주세요.")
        engine.runAndWait()

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    cv2.imshow('Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import pyttsx3
import speech_recognition as sr

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)

cap = cv2.VideoCapture(0)
engine = pyttsx3.init()

def capture_face():
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # 사용자에게 얼굴 사진 촬영 여부 묻기
        engine.say("얼굴을 감지했습니다. 사진을 찍으시겠습니까?")
        engine.runAndWait()

        # 음성 입력 받기
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)
            try:
                user_input = recognizer.recognize_google(audio)
                if user_input.lower() == "yes":
                    # 얼굴 사진 촬영
                    face_img = frame[y:y+h, x:x+w]
                    cv2.imwrite('face_capture.jpg', face_img)
                    engine.say("얼굴 사진을 촬영했습니다.")
                    engine.runAndWait()
            except sr.UnknownValueError:
                pass

    cv2.imshow('Face Detection', frame)

while True:
    capture_face()

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 시간마다 얼굴이 있으면 O 없으면 X
import cv2
import datetime

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)
last_time = None
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    current_time = datetime.datetime.now().strftime('%H:%M:%S')

    if len(faces) > 0:
        if last_time != current_time:
            print(f"{current_time}초\t0")
            last_time = current_time
    else:
        if last_time != current_time:
            print(f"{current_time}초\tX")
            last_time = current_time
    cv2.imshow('Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 머리위 텍스트 
import cv2
from PIL import ImageFont,ImageDraw,Image
import numpy as np 

font= ImageFont.truetype('NanumGothic-Bold.ttf',20)
face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)
img=np.full((200,300,3),(255,255,255),np.uint8)
img=Image.fromarray(img)
draw=ImageDraw.Draw(img)
text='안녕하세요😊'

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        text_x = x + (w // 2) - 100  
        text_y = y - 20  
        cv2.putText(frame, "HELLO WORLD", (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
        draw.text((text_x,text_y),text,font=font,fill=(0,0,0))
        img=np.array(img)
    cv2.imshow('Text on Face', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import csv
import datetime

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)

cap = cv2.VideoCapture(0)

is_frontal_detected = False
frontal_detected_time = None
csv_file = open('frontal_face_times.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Frontal Face Detected Time'])

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    if len(faces) > 0 and not is_frontal_detected:
        is_frontal_detected = True
        frontal_detected_time = datetime.datetime.now()
        csv_writer.writerow([frontal_detected_time.strftime('%Y-%m-%d %H:%M:%S')])

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow('Frontal Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

csv_file.close()
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')


cap = cv2.VideoCapture(0)  

while True:

    ret, frame = cap.read()


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # 눈 찾기
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

    # 영상 출력
    cv2.imshow('Webcam Face Detection', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 작업 완료 후 해제
cap.release()
cv2.destroyAllWindows()
```

```python
#얼굴 꼬북이
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 내가 가진 이미지 로드
img_path = 'pngwing.com.png'
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        # 얼굴 영역에 이미지 삽입
        resized_img = cv2.resize(img, (w, h))
        
        # 알파 채널 사용 여부 확인
        if img.shape[2] == 4:
            alpha_s = resized_img[:, :, 3] / 255.0
            alpha_l = 1.0 - alpha_s
            
            for c in range(0, 3):
                frame[y:y + h, x:x + w, c] = (alpha_s * resized_img[:, :, c] +
                                               alpha_l * frame[y:y + h, x:x + w, c])
        else:
            frame[y:y + h, x:x + w] = resized_img
        
    cv2.imshow('Webcam Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#눈에 라이언

import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

sunglasses_img = cv2.imread('ryan_transparent.png', -1)   
sunglasses_img = cv2.resize(sunglasses_img, (100, 50))  

cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
      
            resized_ryan = cv2.resize(sunglasses_img, (ew, int(eh * 0.7)))
            
           
            x_offset = ex
            y_offset = ey + int(eh * 0.3)
            
            alpha_s = resized_ryan[:, :, 3] / 255.0
            alpha_l = 1.0 - alpha_s
            
            for c in range(0, 3):
                roi_color[y_offset:y_offset + resized_ryan.shape[0], 
                          x_offset:x_offset + resized_ryan.shape[1], c] = (
                              alpha_s * resized_ryan[:, :, c] +
                              alpha_l * roi_color[y_offset:y_offset + resized_ryan.shape[0], 
                                                 x_offset:x_offset + resized_ryan.shape[1], c]
                          )

    cv2.imshow('Webcam Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import csv

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
count = 0  

with open('face_coordinates.csv', mode='w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['x', 'y', 'width', 'height', 'image_path'])  
    
    while True:
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x, y, w, h) in faces:
           
            csv_writer.writerow([x, y, w, h, f'face_{count}.png'])
            
            
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 255, 0), 2)
            
          
            face_img = frame[y:y + h+50, x:x + w+50]
            cv2.imwrite(f'data/face_{count}.png', face_img)
            csv_writer.writerow([x, y, w, h, img_path])
            count += 1
        
        cv2.imshow('Webcam Face Detection', frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import tkinter as tk
from tkinter import filedialog
import cv2
from PIL import Image, ImageTk
import threading

def upload_image():
    file_path = filedialog.askopenfilename()
    img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    img = cv2.resize(img, (100, 100))  # 이미지 크기 조정 (임의로 설정)
    process_image(img)

def process_image(img):
    global face_image
    face_image = img

def detect_face():
    global face_image
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        return
    while True:
        ret, frame = cap.read()
        if ret:
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

            for (x, y, w, h) in faces:
                if face_image is not None:
                    resized_face_image = cv2.resize(face_image, (w, h))
                    alpha_s = resized_face_image[:, :, 3] / 255.0
                    alpha_l = 1.0 - alpha_s

                    for c in range(0, 3):
                        frame[y:y + h, x:x + w, c] = (alpha_s * resized_face_image[:, :, c] +
                                                       alpha_l * frame[y:y + h, x:x + w, c])

            cv2.imshow('Face Detection', frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break
    cap.release()
    cv2.destroyAllWindows()

def start_detection():
    detection_thread = threading.Thread(target=detect_face)
    detection_thread.start()

def camTread():
    color=[]
    cap=cv2.VideoCapture(0)
    panel=None
    if(cap.isOpened()==False):
        print('camera fail')
    while True:
        ret, color=cap.read()
        if(color!=[]):
            cv2.imshow('uvc',color)
            image=cv2.cvtColor(color,cv2.COLOR_BGR2RGB)
            image=Image.fromarray(image)
            image=ImageTk.PhotoImage(image)

            if panel is None:
                panel=tk.Label(image=image)
                panel.image=image
                panel.pack(side='left')
            else:
                panel.configure(image=image)
                panel.image=image

                cv2.waitkey(1)

thread_img=threading.Thread(target=camTread,args=())
thread_img.daemon=True
thread_img.start()

root = tk.Tk()
root.title("Face Recognition")

# # 웹캠 화면
# webcam_panel = tk.Label(root)
# webcam_panel.grid(row=0, column=0)

# # 업로드 버튼
# upload_button = tk.Button(root, text="Upload Image", command=upload_image, width=20, height=2)
# upload_button.grid(row=0, column=1)

# # 창의 위치 설정
# window_position = "+100+100"  # x, y 좌표값
# root.geometry(window_position)

# start_detection()

root.mainloop()

```
