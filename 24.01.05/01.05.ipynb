```python
import cv2

def apply_mosaic(image_path, x, y, width, height):
    # Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
    image = cv2.imread(image_path)

    # Î™®ÏûêÏù¥ÌÅ¨ Ï≤òÎ¶¨Ìï† ÏòÅÏó≠ ÏÑ†ÌÉù
    selected_area = image[y:y+height, x:x+width]

    # Ïù¥ÎØ∏ÏßÄ Ï∂ïÏÜå
    small_area = cv2.resize(selected_area, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_LINEAR)
    # Ïù¥ÎØ∏ÏßÄ ÌôïÎåÄ
    blurred_area = cv2.resize(small_area, (width, height), interpolation=cv2.INTER_NEAREST)

    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóê Î™®ÏûêÏù¥ÌÅ¨ Ï†ÅÏö©
    image[y:y+height, x:x+width] = blurred_area

    # Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
    cv2.imshow('Mosaic Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Ìï®Ïàò Ìò∏Ï∂úÌïòÏó¨ Î™®ÏûêÏù¥ÌÅ¨ Ï†ÅÏö©
apply_mosaic('dog.jpg', 400, 450, 500, 200)
```

```python
import cv2
import pyttsx3
import speech_recognition as sr

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)

cap = cv2.VideoCapture(0)
engine = pyttsx3.init()

def capture_face():
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # ÏÇ¨Ïö©ÏûêÏóêÍ≤å ÏñºÍµ¥ ÏÇ¨ÏßÑ Ï¥¨ÏòÅ Ïó¨Î∂Ä Î¨ªÍ∏∞
        engine.say("ÏñºÍµ¥ÏùÑ Í∞êÏßÄÌñàÏäµÎãàÎã§. ÏÇ¨ÏßÑÏùÑ Ï∞çÏúºÏãúÍ≤†ÏäµÎãàÍπå?")
        engine.runAndWait()

        # ÏùåÏÑ± ÏûÖÎ†• Î∞õÍ∏∞
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)
            try:
                user_input = recognizer.recognize_google(audio)
                if user_input.lower() == "yes":
                    # ÏñºÍµ¥ ÏÇ¨ÏßÑ Ï¥¨ÏòÅ
                    face_img = frame[y:y+h, x:x+w]
                    cv2.imwrite('face_capture.jpg', face_img)
                    engine.say("ÏñºÍµ¥ ÏÇ¨ÏßÑÏùÑ Ï¥¨ÏòÅÌñàÏäµÎãàÎã§.")
                    engine.runAndWait()
            except sr.UnknownValueError:
                pass

    cv2.imshow('Face Detection', frame)

while True:
    capture_face()

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import csv
import datetime

face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_path)

cap = cv2.VideoCapture(0)

is_frontal_detected = False
frontal_detected_time = None
csv_file = open('frontal_face_times.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Frontal Face Detected Time'])

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    if len(faces) > 0 and not is_frontal_detected:
        is_frontal_detected = True
        frontal_detected_time = datetime.datetime.now()
        csv_writer.writerow([frontal_detected_time.strftime('%Y-%m-%d %H:%M:%S')])

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow('Frontal Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

csv_file.close()
cap.release()
cv2.destroyAllWindows()
```

```python
#ÏñºÍµ¥ Íº¨Î∂ÅÏù¥
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# ÎÇ¥Í∞Ä Í∞ÄÏßÑ Ïù¥ÎØ∏ÏßÄ Î°úÎìú
img_path = 'pngwing.com.png'
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        # ÏñºÍµ¥ ÏòÅÏó≠Ïóê Ïù¥ÎØ∏ÏßÄ ÏÇΩÏûÖ
        resized_img = cv2.resize(img, (w, h))
        
        # ÏïåÌåå Ï±ÑÎÑê ÏÇ¨Ïö© Ïó¨Î∂Ä ÌôïÏù∏
        if img.shape[2] == 4:
            alpha_s = resized_img[:, :, 3] / 255.0
            alpha_l = 1.0 - alpha_s
            
            for c in range(0, 3):
                frame[y:y + h, x:x + w, c] = (alpha_s * resized_img[:, :, c] +
                                               alpha_l * frame[y:y + h, x:x + w, c])
        else:
            frame[y:y + h, x:x + w] = resized_img
        
    cv2.imshow('Webcam Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#ÎààÏóê ÎùºÏù¥Ïñ∏

import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

sunglasses_img = cv2.imread('ryan_transparent.png', -1)   
sunglasses_img = cv2.resize(sunglasses_img, (100, 50))  

cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
      
            resized_ryan = cv2.resize(sunglasses_img, (ew, int(eh * 0.7)))
            
           
            x_offset = ex
            y_offset = ey + int(eh * 0.3)
            
            alpha_s = resized_ryan[:, :, 3] / 255.0
            alpha_l = 1.0 - alpha_s
            
            for c in range(0, 3):
                roi_color[y_offset:y_offset + resized_ryan.shape[0], 
                          x_offset:x_offset + resized_ryan.shape[1], c] = (
                              alpha_s * resized_ryan[:, :, c] +
                              alpha_l * roi_color[y_offset:y_offset + resized_ryan.shape[0], 
                                                 x_offset:x_offset + resized_ryan.shape[1], c]
                          )

    cv2.imshow('Webcam Face Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import tkinter as tk
from tkinter import filedialog
import cv2
from PIL import Image, ImageTk
import threading

def upload_image():
    file_path = filedialog.askopenfilename()
    img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    img = cv2.resize(img, (100, 100))  # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï°∞Ï†ï (ÏûÑÏùòÎ°ú ÏÑ§Ï†ï)
    process_image(img)

def process_image(img):
    global face_image
    face_image = img

def detect_face():
    global face_image
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        return
    while True:
        ret, frame = cap.read()
        if ret:
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

            for (x, y, w, h) in faces:
                if face_image is not None:
                    resized_face_image = cv2.resize(face_image, (w, h))
                    alpha_s = resized_face_image[:, :, 3] / 255.0
                    alpha_l = 1.0 - alpha_s

                    for c in range(0, 3):
                        frame[y:y + h, x:x + w, c] = (alpha_s * resized_face_image[:, :, c] +
                                                       alpha_l * frame[y:y + h, x:x + w, c])

            cv2.imshow('Face Detection', frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break
    cap.release()
    cv2.destroyAllWindows()

def start_detection():
    detection_thread = threading.Thread(target=detect_face)
    detection_thread.start()

def camTread():
    color=[]
    cap=cv2.VideoCapture(0)
    panel=None
    if(cap.isOpened()==False):
        print('camera fail')
    while True:
        ret, color=cap.read()
        if(color!=[]):
            cv2.imshow('uvc',color)
            image=cv2.cvtColor(color,cv2.COLOR_BGR2RGB)
            image=Image.fromarray(image)
            image=ImageTk.PhotoImage(image)

            if panel is None:
                panel=tk.Label(image=image)
                panel.image=image
                panel.pack(side='left')
            else:
                panel.configure(image=image)
                panel.image=image

                cv2.waitkey(1)

thread_img=threading.Thread(target=camTread,args=())
thread_img.daemon=True
thread_img.start()

root = tk.Tk()
root.title("Face Recognition")

# # ÏõπÏ∫† ÌôîÎ©¥
# webcam_panel = tk.Label(root)
# webcam_panel.grid(row=0, column=0)

# # ÏóÖÎ°úÎìú Î≤ÑÌäº
# upload_button = tk.Button(root, text="Upload Image", command=upload_image, width=20, height=2)
# upload_button.grid(row=0, column=1)

# # Ï∞ΩÏùò ÏúÑÏπò ÏÑ§Ï†ï
# window_position = "+100+100"  # x, y Ï¢åÌëúÍ∞í
# root.geometry(window_position)

# start_detection()

root.mainloop()

```
