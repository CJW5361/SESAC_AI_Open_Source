```python
def calculate_statistics(df_group):
    count = df_group['publisher'].count()
    mean_price = df_group['discount'].mean()
    median_price = df_group['discount'].median()
    std_price = df_group['discount'].std()
    min_price = df_group['discount'].min()
    max_price = df_group['discount'].max()
    date=df_group['year'].mean()

    return pd.Series({
        '출판 횟수': count,
        '평균 가격': mean_price,
        '중앙값 가격': median_price,
        '표준 편차 가격': std_price,
        '최소 가격': min_price,
        '최대 가격': max_price,
        '날짜':date
    })

def calculate_publishers_statistics(df):
    publishers_statistics = df.groupby('publisher').apply(calculate_statistics)
    return publishers_statistics

# 출판사별 도서 통계 계산
publishers_statistics = calculate_publishers_statistics(df)

# 결과 출력
publishers_statistics
```

```python
import pandas as pd
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
# 가정: df는 지정된 열을 가진 데이터프레임입니다.
# 없다면 CSV 파일이나 다른 소스에서 데이터프레임을 로드할 수 있습니다.
# 예: df = pd.read_csv('your_data.csv')

# 형태소 분석을 위한 Okt 객체 생성
okt = Okt()


df['text'] = df['title'] + ' ' + df['description']

# 'text' 열의 데이터를 년도별로 그룹화하여 형태소 분석 및 명사 추출
def extract_nouns(text):
    nouns = okt.nouns(text)
    return nouns

nouns_by_year = df.groupby(df['pubdate'].dt.year)['text'].apply(lambda x: ' '.join(x)).apply(extract_nouns)

# 각 년도별 명사 빈도수 카운트
nouns_counts_by_year = nouns_by_year.apply(lambda x: Counter(x))

# 결과 확인
for year, nouns_counts in nouns_counts_by_year.items():
    print(f"Year {year} - Most common nouns: {nouns_counts.most_common(5)}")
```

```python
for year, nouns_counts in nouns_counts_by_year.items():
    wordcloud = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', background_color='white').generate_from_frequencies(nouns_counts)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for Year {year}')
    plt.axis('off')
    plt.show()
```
