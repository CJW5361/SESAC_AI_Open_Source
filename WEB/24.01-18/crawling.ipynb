```python
import requests
from bs4 import BeautifulSoup

# 이전에 어떻게 soup을 생성했는지에 따라서 코드를 조절해야 할 수 있습니다.
# 아래는 간단한 예시로 작성한 것입니다.
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, 'html.parser')

carimglist = []

for i in range(0, 20):
    # 직접 nth-child를 사용하지 않고, BeautifulSoup을 사용하여 각 li 요소를 선택합니다.
    selector = f'div.wrap-thumb-list > ul > li:nth-child({i}) > div > div.mode-cell.thumb > a > img'
    carimg = soup.select(selector)

    # 만약 이미지를 찾았다면 리스트에 추가합니다.
    if carimg:
        carimglist.append(carimg[0].get('src'))

print(carimglist)
```
