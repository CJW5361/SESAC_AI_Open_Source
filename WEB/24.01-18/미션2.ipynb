```python
import requests

# 네이버쇼핑 API 엔드포인트 및 API 키 설정
api_endpoint = "https://openapi.naver.com/v1/search/shop.json"
client_key = ""
client_secret=''
# API 호출을 위한 파라미터 설정
total_results = 100

# 각 요청당 가져올 결과의 개수
results_per_request = 100

# 가져온 결과를 저장할 리스트
data = []
all_results = []
# 여러 번의 요청을 통해 결과 가져오기
for start in range(1, total_results + 1, results_per_request):
    # API 호출을 위한 파라미터 설정
    params = {
        'query': '노트북 삼성전자',
        'price_max': 1000000,
        'start': start,
        'display': results_per_request,
    }

    # API 호출 및 응답 받기
    headers = {'X-Naver-Client-Id': client_key, 'X-Naver-Client-Secret': client_secret}
    response = requests.get(api_endpoint, params=params, headers=headers)
    result = response.json()

# 결과에서 필요한 정보 추출 및 처리
    if 'items' in result:
            for item in result['items']:
                # 가격이 100만원 이하인 제품만 출력
                if int(item['lprice']) <= 1000000:
                    print(f"제품명: {item['title']}, 가격: {item['lprice']}원")
                    all_results.append(item)
    else:
            print("검색 결과가 없습니다.")

# 만약에 모든 결과를 저장하고 싶다면, 이후에 all_results 리스트 활용 가능
```

```python
import pandas as pd
import requests

# 네이버쇼핑 API 엔드포인트 및 API 키 설정
api_endpoint = "https://openapi.naver.com/v1/search/shop.json"
api_key = ""

# 검색어 설정
query = "LG 스마트TV"

# API 호출을 위한 파라미터 설정
params = {
    'query': query,
    'display': 100,  # 100개까지 결과를 가져옴
}

# API 호출 및 응답 받기
headers = {'X-Naver-Client-Id': api_key, 'X-Naver-Client-Secret': ''}
response = requests.get(api_endpoint, params=params, headers=headers)
result = response.json()

# 결과에서 필요한 정보 추출 및 처리
data = []

if 'items' in result:
    for item in result['items']:
        product_info = {
            '제품명': item['title'],
            '가격': int(item['lprice']),
            '쇼핑몰': item['mallName'],
            '링크': item['link'],
        }
        data.append(product_info)

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 같은 제품별로 가장 최저가인 쇼핑몰 및 링크 찾기
min_prices = df.groupby('제품명')['가격'].idxmin()
result_df = df.loc[min_prices]

# 결과 출력
display(result_df[['제품명', '가격', '쇼핑몰', '링크']])
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 네이버쇼핑 API 엔드포인트 및 API 키 설정
api_endpoint = "https://openapi.naver.com/v1/search/shop.json"
api_key = ""

# 검색어 설정
query = "LG 스마트TV"

# API 호출을 위한 파라미터 설정
params = {
    'query': query,
    'display': 100,  # 100개까지 결과를 가져옴
}

# API 호출 및 응답 받기
headers = {'X-Naver-Client-Id': api_key, 'X-Naver-Client-Secret': ''}
response = requests.get(api_endpoint, params=params, headers=headers)
result = response.json()

# 결과에서 필요한 정보 추출 및 처리
data = []

if 'items' in result:
    for item in result['items']:
        # HTML 태그 제거
        soup = BeautifulSoup(item['title'], 'html.parser')
        product_name = soup.get_text()

        product_info = {
            '제품명': product_name,
            '가격': int(item['lprice']),
            '쇼핑몰': item['mallName'],
            '링크': item['link'],
        }
        data.append(product_info)

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 중복 제거
df.drop_duplicates(subset='제품명', keep='first', inplace=True)
df=df[['제품명', '가격', '쇼핑몰', '링크']]
# 결과 출력
display(df)

```

```python
import requests
import os

# Google Custom Search API 정보 설정
api_key = ""
cx = ""

# 검색어 설정
search_query = "박은빈"

# Google Custom Search API 호출을 위한 URL 설정
api_url = f"https://www.googleapis.com/customsearch/v1?q={search_query}&key={api_key}&cx={cx}&searchType=image"

# API 호출 및 응답 받기
response = requests.get(api_url)
data = response.json()

# 이미지 다운로드
if 'items' in data:
    for item in data['items']:
        image_url = item['link']
        image_data = requests.get(image_url).content

        # 이미지 저장
        with open(os.path.basename(image_url), 'wb') as f:
            f.write(image_data)
            print(f"Image '{os.path.basename(image_url)}' downloaded successfully.")
else:
    print("No image found.")
```

```python
import googlemaps
from datetime import datetime
my_key=''
maps=googlemaps.Client(key=my_key)

import pandas as pd

df=pd.read_excel('address.xlsx')
df=df.rename(columns={'서울 강남구 선릉로 836 삼원빌딩 1~2층' : '장소'})
df
```

```python
data=df['장소']
print(data)
```

```python
import time
lat=[]
lng=[]

i=0
t1=time.time()
for address in df['장소']:
  i=i+1
  try:
    geo_location=maps.geocode(address)[0].get('geometry')
    lat.append(geo_location['location']['lat'])
    lng.append(geo_location['location']['lng'])

  except:
    lat.append('')
    lng.append('')
    print('%d번 인덱스 에러'%(i))

print(time.time() -t1)
```

```python
df2=pd.DataFrame({'주소':df['장소'],'위도':lat,'경도':lng})

df2
```
